### Plots and subplots

Perfect, we now have matplotlib installed in our conda environment and imported in the jupyter notebook. To use matplotlib's built-in functions, we will call the commands using `plt.method`.

Before we move onto real data analysis, let's get some background on how matplotlib organizes graphs:

```python
#Some very basic data to plot:
x = [1,2,3,4,5,6,7,8,9]
y = [num**2 for num in x]
print(x)
print(y)
```

```python
#The plot function takes as input an x and y and gets as output a line graph
plt.plot(x,y)
```

```python
#there are tons of customization options, we won't go over all of them but feel free to explore:
plt.plot(x,y, color = "red", lw = 3, ls = "--")
```

```python
#We can also add a title and labels to the axis
plt.plot(x,y, color = "red", lw = 3, ls = "--")
plt.xlabel("x sequence")
plt.ylabel("x squared")
plt.title("X vs Y")
```

```python
#Matplotlib can also plot different graphs on the same canvas
fig, ax = plt.subplots(1,1, figsize=(4,2)) #one subplot in a canvas size of 4 by 2
```

```python
fig, ax = plt.subplots(1,2, figsize=(4,2)) #the size specifies the whole canvas size, so we might want to make it larger
plt.tight_layout() #helps position the plots without overlap
```

```python
fig, ax = plt.subplots(2,2, figsize=(4,4)) #the size specifies the whole canvas size, so we might want to make it larger
ax1 = ax[0][1]
#to now plot on one of the empty graphs we have created:
ax1.plot(x,y)
plt.tight_layout()
```

### Import dataset to work with
Let's start by importing a dataset to work on. We will re-use the dataset from the session 3 pre-exercises, "DatasetAfricaMalaria.csv"

```python
import pandas as pd
df = pd.read_csv("DatasetAfricaMalaria.csv")
df.head()
```

## Categorical plots

### Barplots

```python
#Malaria incidence in year 2007
y2007 = df[df["Year"]==2007] #create a dataframe with the values from year 2007 only
x = y2007["Country Code"].tolist() #the x values is a list of the countries
y = y2007["Incidence of malaria (per 1,000 population at risk)"] #the y values, or height of the bar, is the malaria incidence

plt.bar(x=x, height = y)
```

```python
#for example, we can try to spread the figure size so all countries fit
fig, ax = plt.subplots(1,1, figsize=(25,5))
ax.bar(x=y2007["Country Code"], height = y2007["Incidence of malaria (per 1,000 population at risk)"])
```

By calling the .bar function off the same ax, we stack the graphs on top of each other

```python
#stacked barplots
#Malaria incidence in year 2007
y2007 = df[df["Year"]==2007] #create a dataframe with the values from year 2007 only
x = y2007["Country Code"].tolist() #the x values is a list of the countries
y = y2007["Incidence of malaria (per 1,000 population at risk)"] #the y values, or height of the bar, is the malaria incidence

#Malaria incidence in year 2007
y2008 = df[df["Year"]==2007] #create a dataframe with the values from year 2007 only
x2 = y2008["Country Code"].tolist() #the x values is a list of the countries
y2 = y2008["Incidence of malaria (per 1,000 population at risk)"] #the y values, or height of the bar, is the malaria incidence

fig, ax = plt.subplots(1,1, figsize=(25,5))


ax.bar(x=x, height = y, color = "Blue", label = "2007")
ax.bar(x=x2, height = y2, color = "Orange", bottom=y, label ="2008")
plt.legend()
```

### Boxplot

```python
ago = df[df["Country Code"] ==  "AGO"]

plt.boxplot(ago["Malaria cases reported"])
plt.ylabel("Malaria cases (x1e6)")
plt.title("Malaria cases in Angola between 2007 - 2017")
plt.show()
```

```python
ago = df[df["Country Code"] ==  "AGO"]
ken = df[df["Country Code"] == "KEN"]
bfa = df[df["Country Code"] == "BFA"]

boxlist = [ago["Malaria cases reported"], ken["Malaria cases reported"], bfa["Malaria cases reported"]]
lab = ["Angola", "Kenya", "BurkinaFaso"]

plt.boxplot(boxlist)
plt.ylabel("Malaria cases (x1e6)")
plt.title("Malaria cases between 2007 - 2017")
plt.xticks(ticks = [1,2,3], labels = lab)
plt.show()
```

### Violinplot

```python
ago = df[df["Country Code"] ==  "AGO"]
ken = df[df["Country Code"] == "KEN"]
bfa = df[df["Country Code"] == "BFA"]

boxlist = [ago["Malaria cases reported"], ken["Malaria cases reported"], bfa["Malaria cases reported"]]
lab = ["Angola", "Kenya", "BurkinaFaso"]

plt.violinplot(boxlist)
plt.ylabel("Malaria cases (x1e6)")
plt.title("Malaria cases between 2007 - 2017")
plt.xticks(ticks = [1,2,3], labels = lab)
plt.show()
```

## Distribution plots

### Lineplots

```python
#let's start by checking the number of malaria cases over the years
bfa = df[df["Country Code"] == "BFA"]

plt.plot(bfa["Year"], bfa["Malaria cases reported"], lw=2, color = "green")
plt.title("Reported Malaria cases in Burkina Faso")
plt.xlabel("years")
plt.ylabel("Malaria cases 1e7")
```

```python
countries = df["Country Code"].tolist()
countries[:10]
```

```python
for country in countries[:10]:
    subset = df[df["Country Code"]==country]
    plt.plot(subset["Year"], subset["Malaria cases reported"], lw=1)
    
plt.legend(df["Country Code"])
```

```python
#we can use the .groupby method to simplify the analysis

years = df.groupby("Year").sum()
years.head()
```

```python
plt.plot(years["Incidence of malaria (per 1,000 population at risk)"], marker=".", ms=10)
plt.xlabel("years")
plt.ylabel("Malaria incidence")
plt.title("Malaria incidence from 2007 to 2017")
```

### Scatter plot

Scatter plots are very useful when we want to correlate variables

```python
incidence = df["Incidence of malaria (per 1,000 population at risk)"]
report = df["Malaria cases reported"]

plt.scatter(report, incidence)
plt.xlabel("Malaria Cases Reported")
plt.ylabel("Malaria Incidence (per 1000 habs)")
```

```python
plt.scatter(df["Use of insecticide-treated bed nets (% of under-5 population)"], df["Incidence of malaria (per 1,000 population at risk)"] )
plt.xlabel("Insecticide treated nets %")
plt.ylabel("Malaria Incidence")
```

### Histograms

```python
plt.hist(df["Use of insecticide-treated bed nets (% of under-5 population)"])
```

# Geographical maps
This is a very specific function to plot data in a geographical map, but since we have used the malaria dataset, we can explore how to plot the data for each country.

You can read more about Choropleth maps and the plotly library here: https://plotly.com/python/choropleth-maps/#reference 

```python
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df2017 = df[df["Year"]==2017]

fig = go.Figure(data=go.Choropleth(
    locations=df2017["Country Code"],
    z = df2017["Malaria cases reported"],
    text=df2017["Country Name"],
    autocolorscale=False,
    reversescale=False,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_tickprefix = None,
    colorbar_title = 'Malaria cases reported',
))
fig.show()
```
